```python
data = pd.read_csv("aw_fb_data.csv", sep=",", index_col=0)
```

```python
data = data.drop("X1", axis=1)
```

```python
filter_data = pd.get_dummies(data)
```

```python
filter_data.describe().T
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
std_scaler.fit(filter_data)
x_data_norm=std_scaler.transform(filter_data)

# A ver que hizo!?
pd.DataFrame(x_data_norm).describe()
```

```python
from sklearn.cluster import KMeans

cl_model = KMeans(
    n_clusters=3,     # el parámetro importante!
    init='k-means++', # k-means++ acelera la convergencia, respecto de random
    max_iter=100,
    random_state=0,
    n_init='auto'     # Para evitar warnings más adelante
)

cl_model.fit(x_data_norm)
```

```python
pred_cl = cl_model.predict(x_data_norm)

# Número de cluster:
pred_cl
```

```python
cl_model
```

```python
from copy import deepcopy
```

```python
from yellowbrick.cluster.elbow import kelbow_visualizer

cl_model2=deepcopy(cl_model) # ojo

kelbow_visualizer(cl_model2, x_data_norm, k=(2,30), locate_elbow=True)
```

```python
cluster_var='cluster' # definan nombre de var de clusteringa
x_data_cl=filter_data.copy() # hagan copia de los datos de TR para no perderlos antes de normalizar
x_data_cl[cluster_var]=cl_model.predict(x_data_norm) # ojo aca! el predict es de los datos normalizados
```

```python
x_data_cl.head()
```

```python
from funpymodeling.model_validation import coord_plot
```

```python
d_orig, d_transf = coord_plot(x_data_cl, cluster_var)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear una tabla pivote con los valores promediados por cluster
pivot_table = x_data_cl.groupby('cluster').mean()

# Crear un mapa de calor
plt.figure(figsize=(12, 8))  #Ajusta el tamaño del mapa de calor según tus necesidades
sns.heatmap(pivot_table, cmap='YlGnBu')  # Cambia el mapa de colores si lo deseas
plt.title("Heatmap of Cluster Features")
plt.show()
```

```python
# Ese 4 significa que cada cluster va a tener como mínimo 4 casos
model_hdb2 = hdbscan.HDBSCAN(min_cluster_size=80).fit(filter_data)
```

```python
freq_tbl(model_hdb2.labels_.astype('str'))
```

```python
cluster_var='cluster' # definan nombre de var de clustering
x_data_cl=filter_data.copy() # hagan copia de los datos de TR para no perderlos
x_data_cl[cluster_var]=model_hdb2.labels_ # ojo aca! el predict es de los datos normalizados
```

```python
x_data_cl.head()
```

Vemos que a traves de estos datos los usarios solamente interactuan con applewacht, son usuarios de edad joven, los usuarios en este conjunto de datos parecen ser personas jóvenes (promedio de 20 años) y en su mayoría hombres (género 1). Tienen una altura y peso consistentes (168 cm y 65.4 kg respectivamente). A medida que aumenta el tiempo, observamos un aumento constante en la cantidad de pasos, y también un aumento en la frecuencia cardíaca y el consumo de calorías. Esto podría indicar que estos usuarios están realizando actividades físicas progresivamente más intensas y enérgicas.
